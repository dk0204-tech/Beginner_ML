```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv(
    "dataset/movies.csv"
)
```

```python
features = [
    
    'genres',
    
    'keywords',
    
    'overview',
    
    'cast',
    
    'director'
]
```

```python
for feature in features:
    
    df[feature] = df[feature].fillna('')
```

```python
combined_features = (
    
    df['genres'] + ' ' +
    
    df['keywords'] + ' ' +
    
    df['overview'] + ' ' +
    
    df['cast'] + ' ' +
    
    df['director']
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')

feature_vectors = vectorizer.fit_transform(
    combined_features
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(
    feature_vectors
)
```

```python
def recommend_movies(movie_name):
    
    movie_list = df['title'].tolist()
    
    import difflib
    
    close_match = difflib.get_close_matches(
        movie_name,
        movie_list
    )[0]
    
    movie_index = df[
        df.title == close_match
    ]['index'].values[0]
    
    similarity_score = list(
        enumerate(
            similarity[movie_index]
        )
    )
    
    sorted_movies = sorted(
        similarity_score,
        
        key=lambda x:x[1],
        
        reverse=True
    )
    
    print(f"\nMovies similar to {close_match}:\n")
    
    i = 1
    
    for movie in sorted_movies:
        
        index = movie[0]
        
        title = df[df.index == index]['title'].values[0]
        
        if i < 11:
            
            print(i, ".", title)
            
            i += 1
```

```python
recommend_movies("Iron")
```
